```python
# ì§„í–‰ê²½ê³¼ í‘œê¸°.
# !pip install geopy
import pandas as pd
import time
import csv
import os
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from concurrent.futures import ThreadPoolExecutor, as_completed

# Nominatim API ì´ˆê¸°í™”
geolocator = Nominatim(user_agent="geo_converter")

# CSV íŒŒì¼ ë¡œë“œ
df = pd.read_csv("í•œêµ­ì „ë ¥ê³µì‚¬_ë¶€ì‚°ì‹œ ì „ê¸°ì°¨ ì¶©ì „ì†Œ ì¶©ì „ëŸ‰_20220331.csv", encoding="utf-8")

# "ì£¼ì†Œ" í•„ë“œ ì„ íƒ í›„ ë¦¬ìŠ¤íŠ¸ ë³€í™˜
addresses = df["ì£¼ì†Œ"].dropna().tolist()

# ì „ì²´ ì£¼ì†Œ ê°œìˆ˜
total_count = len(addresses)
progress_count = 0  # ì§„í–‰ ê°œìˆ˜ ì¶”ì 

# ê²°ê³¼ ì €ì¥ CSV íŒŒì¼ ê²½ë¡œ
output_csv = "converted_addresses.csv"

# CSV íŒŒì¼ì´ ì—†ëŠ” ê²½ìš° í—¤ë” ì¶”ê°€
if not os.path.exists(output_csv):
    with open(output_csv, mode="w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(["ì£¼ì†Œ", "ìœ„ë„", "ê²½ë„"])  # í—¤ë” ì¶”ê°€

# ì£¼ì†Œ ë³€í™˜ í•¨ìˆ˜ (ì˜ˆì™¸ ì²˜ë¦¬ + 1.5ì´ˆ ëŒ€ê¸° + ì§„í–‰ ìƒíƒœ ì—…ë°ì´íŠ¸)
def get_lat_lon(address):
    global progress_count  # ì§„í–‰ ê°œìˆ˜ ì—…ë°ì´íŠ¸ë¥¼ ìœ„í•´ ì „ì—­ ë³€ìˆ˜ ì‚¬ìš©

    try:
        location = geolocator.geocode(address, timeout=10)  # 10ì´ˆ ëŒ€ê¸°
        if location:
            lat, lon = location.latitude, location.longitude
        else:
            lat, lon = None, None
    except GeocoderTimedOut:
        print(f"â³ ì‹œê°„ ì´ˆê³¼: {address}, ë‹¤ì‹œ ì‹œë„ ì¤‘...")
        time.sleep(2)
        return get_lat_lon(address)  # ì¬ì‹œë„

    # ë³€í™˜ ì™„ë£Œ í›„ CSVì— í•œ ì¤„ì”© ì¶”ê°€ ì €ì¥
    with open(output_csv, mode="a", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow([address, lat, lon])

    # ì§„í–‰ ê°œìˆ˜ ì—…ë°ì´íŠ¸
    progress_count += 1
    progress_percent = (progress_count / total_count) * 100
    print(f"âœ… ì§„í–‰ ìƒíƒœ: {progress_count}/{total_count} ì™„ë£Œ ({progress_percent:.2f}%)")

    time.sleep(1.5)  # 1.5ì´ˆ ëŒ€ê¸° (ì†ë„ ì œí•œ ë°©ì§€)
    return address, lat, lon

# ë©€í‹°ìŠ¤ë ˆë”© ì‹¤í–‰ (ìµœëŒ€ 5ê°œ ë™ì‹œ ì‹¤í–‰)
with ThreadPoolExecutor(max_workers=5) as executor:
    future_to_address = {executor.submit(get_lat_lon, addr): addr for addr in addresses}

    for future in as_completed(future_to_address):
        try:
            future.result()  # ê° ë³€í™˜ ì‘ì—… ì‹¤í–‰
        except Exception as e:
            print(f"âš ï¸ ì˜¤ë¥˜ ë°œìƒ: {e}")

print("ğŸ‰ ëª¨ë“  ì£¼ì†Œ ë³€í™˜ ì™„ë£Œ! ë³€í™˜ëœ ë°ì´í„°ê°€ 'converted_addresses.csv'ì— ì‹¤ì‹œê°„ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
